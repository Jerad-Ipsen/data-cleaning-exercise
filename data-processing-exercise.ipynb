```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('ml-03-data-processing-songs-dataset.csv')
print(df.shape)
df
```

```python
print(df.info())
df.hist(figsize = (10, 10))
df.describe()
```

```python
# 1

df['Length (Duration)'] = df['Length (Duration)'].\
    str.replace(',','').astype('float')
df.info()
```

```python
# 2

df_dropped_columns = df.drop(columns = ['Danceability', 'Valence'])
print(f'{df_dropped_columns.isnull().sum(axis=1).sum()}\n')
df_dropped_columns.info()

# Dropping rows? (finding rows with 50% null)
```

```python
# 3

df_dropped_columns.describe()

for column in df_dropped_columns:
    if(pd.api.types.is_numeric_dtype(df_dropped_columns[column])):
        mean = df_dropped_columns[column].mean()
        std = df_dropped_columns[column].std()
        new_df = df_dropped_columns[(df_dropped_columns[column] < (mean - 3 * std)) | (df_dropped_columns[column] > (mean + 3 * std))]
        print(f'Column: {column}')
        print(f'Mean: {mean}, Std: {std}')
        print(f'Outlier count: {new_df.shape[0]}')
        plt.boxplot(new_df[column])
        plt.show()

df_dropped_columns['Year'].loc[df_dropped_columns['Year'] < 1000] = 1992
df_dropped_columns[df_dropped_columns['Year'] < 1000]
```

```python
# 4 (?)



df_fill_title = df_dropped_columns['Title'].fillna(df_dropped_columns['Title'].mode().iloc[1])
df_fill_artist = df_fill_title['Artist'].fillna(df_fill_title['Artist'].mode().iloc[2])
df_fill_genre = df_fill_artist['Top Genre'].fillna(df_fill_artist['Top Genre'].mode().iloc[3])
df_fill_year = df_fill_genre['Year'].fillna(df_fill_genre['Year'].mean().iloc[4])
df_fill_bpm = df_fill_year['Beats Per Minute (BPM)'].fillna(df_fill_year['Beats Per Minute (BPM)'].mean().iloc[6])
df_fill_energy = df_fill_bpm['Energy'].fillna(df_fill_bpm['Energy'].mean().iloc[7])
df_fill_loudness = df_fill_energy['Loudness (dB)'].fillna(df_fill_energy['Loudness (dB)'].mean().iloc[8])
df_fill_liveness = df_fill_loudness['Liveness'].fillna(df_fill_loudness['Liveness'].mean().iloc[9])
df_fill_length = df_fill_liveness['Length (Duration)'].fillna(df_fill_liveness['Length (Duration)'].mean().iloc[10])
df_fill_acousticness = df_fill_length['Acousticness'].fillna(df_fill_length['Acousticness'].mean().iloc[11])
df_fill_speechiness = df_fill_acousticness['Speechiness'].fillna(df_fill_acousticness['Speechiness'].mean().iloc[12])
df_fill_na = df_fill_speechiness['Popularity'].fillna(df_fill_speechiness['Popularity'].mean().iloc[13])

df_fill_na.info()
```
